# Corelation co-efficient relaition and covarience analysis

```python
import pandas as pd
import numpy as ny
import seaborn as sns
import matplotlib.pyplot as plt
import math
import statistics as st
% matplotlib in line
```

```python
std = pd.read_csv('student-mat.csv')
std.head(10)
```

```python
fig_dims = (15,10)
fig, ax = plt.subplots(figsize=fig_dims)
sns.heatmap(data=std.corr(),annot=True).set_title('student-mat')
```

```python
columns_of_interest2 = [ 'studytime','freetime']
```

```python
subset_data = std[columns_of_interest2]
```

```python
correlation_matrix = subset_data.corr()
```

```python
covariance_matrix = subset_data.cov()
```

```python
print("Correlation Matrix:")
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix,cmap ="YlGnBu")
plt.show()
```

```python
print("\nCovariance Matrix:")
print(covariance_matrix)
```

```python
sns.heatmap(covariance_matrix,cmap ="coolwarm")
plt.show()
```
